**Generate heatmaps for each player in each match (各試合・各選手ごとにヒートマップを生成）**

```python
# Import necessary libraries
# 必要なライブラリのインポート
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm  # Progress bar for loops / ループ処理の進捗表示

# Define input folder containing event JSON files
# イベントJSONファイルが保存されている入力フォルダを定義
folder_path = "/content/drive/MyDrive/MIT_Sloan/fifa_wc_2022_events"

# Define output folder for saving heatmaps
# ヒートマップ画像の保存先フォルダを定義
output_dir = "/content/drive/MyDrive/MIT_Sloan/heatmaps_all"
os.makedirs(output_dir, exist_ok=True)  # Create folder if it doesn't exist / フォルダがなければ作成

# Load all event data from JSON files
# 全試合のイベントデータを読み込み
events_data = []
for filename in tqdm(os.listdir(folder_path)):
    if filename.endswith(".json"):
        match_id = filename.replace(".json", "")
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            events = json.load(f)
            for event in events:
                event["match_id"] = match_id  # Add match ID to each event / 各イベントにmatch_idを追加
                if "player" in event:
                    event["player_id"] = event["player"]["id"]  # Extract player ID / player_idを抽出
                else:
                    event["player_id"] = None
                events_data.append(event)

# Convert event data to DataFrame and filter for passes
# イベントデータをDataFrameに変換し、パスイベントのみ抽出
df = pd.DataFrame(events_data)
df = df[df['type'].apply(lambda x: x['name'] == 'Pass')]

# Generate heatmaps for each player in each match
# 各試合・各選手ごとにヒートマップを生成（上下反転処理付き）
for (match_id, player_id), group in df.groupby(['match_id', 'player_id']):
    if player_id is None or len(group) < 5:
        continue  # Skip if player ID is missing or too few passes / player_idがない、またはパスが少ない場合はスキップ

    # Extract and transform coordinates (flip vertically)
    # 座標を抽出し、上下反転（y座標）を適用
    coords = group['location'].dropna().tolist()
    x_coords = [c[0] for c in coords if isinstance(c, list) and len(c) == 2]
    y_coords = [80 - c[1] for c in coords if isinstance(c, list) and len(c) == 2]  # Flip vertically / 上下反転

    if len(y_coords) == 0:
        continue  # Skip if no valid coordinates / 有効な座標がない場合はスキップ

    # Create heatmap without axes or labels
    # 軸・目盛り・ラベルなしのヒートマップを描画
    plt.figure(figsize=(6, 4))
    sns.kdeplot(x=x_coords, y=y_coords, shade=True, cmap="viridis", bw_adjust=0.5)
    plt.axis('off')
    plt.xticks([])
    plt.yticks([])
    plt.title('')
    plt.xlabel('')
    plt.ylabel('')
    plt.xlim(0, 120)
    plt.ylim(0, 80)

    # Save heatmap image
    # ヒートマップ画像を保存
    filename = f'pass_heatmap_match_{match_id}_player_{player_id}.png'
    filepath = os.path.join(output_dir, filename)
    plt.savefig(filepath, bbox_inches='tight', pad_inches=0)
    plt.close()

    print(f"Saved: {filepath}")  # 保存完了メッセージ
```
